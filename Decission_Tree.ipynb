The name of the dataset i'm using is HR-Employee-Attrition, it was found at (https://www.kaggle.com/datasets/rishikeshkonapure/hr-analytics-prediction), and (https://www.kaggle.com/datasets/ppb00x/country-gdp)This dataset. have  information for an employee, such as salary income, years of service , family, and gdp of the country .Which are the factors that modeling the human behavior in relation to savings and withdrawal patterns in the TwoPot system

```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("C:/Users/Admin/Downloads/archive (3)/HR-Employee-Attrition.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['MonthlyIncome'] = df['MonthlyIncome'].astype(float)
df.head()
```

```python
df['Attrition'] = df['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)
df.head()
```

```python

data= df.drop(columns=['EmployeeCount', 'EmployeeNumber', 'Over18'])
df.head()
```

```python

df['RecentlyPromoted'] = df['YearsSinceLastPromotion'].apply(lambda x: 1 if x <= 5 else 0)
df['JobSatisfaction_Salary'] = df['JobSatisfaction'] * df['MonthlyIncome']
df['WorkLifeBalance_Satisfaction'] = df['WorkLifeBalance'] * df['JobSatisfaction']
```

```python
data = pd.get_dummies(data, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

X = data.drop('Attrition', axis=1)
y = data['Attrition']  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=45)

print(f'Training set shape: {X_train.shape}, Training labels shape: {y_train.shape}')
print(f'Testing set shape: {X_test.shape}, Testing labels shape: {y_test.shape}')
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix


dt_model = DecisionTreeClassifier(random_state=45)


dt_model.fit(X_train, y_train)
y_pred = dt_model.predict(X_test)


print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
new_employee = [[35, 10, 60000, 1, 0]]
predicted_withdrawal_dt = dt_model.predict(new_employee)
if predicted_withdrawal_dt[0] == 1:
    print("Decision Tree: The employee is likely to withdraw funds.")
else:
    print("Decision Tree: The employee is unlikely to withdraw funds.")
```
